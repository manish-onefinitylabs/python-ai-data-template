### Exercise 2: Basic Seaborn Bar Plot

```python
# Task: Create a bar plot using seaborn to visualize the number of sales (in thousands) for a set of
# products: 'A', 'B', 'C', 'D', and 'E' with values [23, 75, 57, 90, 65].
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Data
products = ['A', 'B', 'C', 'D', 'E']
sales = [23, 75, 57, 90, 65]

# Create a bar plot
sns.set(style="whitegrid")  # Set Seaborn style
plt.figure(figsize=(8, 5))  # Set figure size
sns.barplot(x=products, y=sales, palette='Blues')

# Adding titles and labels
plt.title("Sales of Products", fontsize=14)
plt.xlabel("Products", fontsize=12)
#plt.ylabel("Sales (in Thousands)", fontsize=12)

# Show the plot
#plt.show()
```
